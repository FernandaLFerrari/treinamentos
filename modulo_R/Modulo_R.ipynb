### Condicionais e Repetição


* Nas condicionais precisamos dar um teste lógico para o R validar e depois dizer para ele o que fazer caso a condição seja VERDADEIRA e o que fazer caso a condição seja FALSA.

**if (teste_logico) {**
  
          comandos que serão rodados
          
          quando a condição for verdadeira
  
**}**


### Variação

**if (teste_logico) {**
  
          comandos que serão rodados
          
          quando a condição for verdadeira

**}else{**
  
          comandos que serão rodados
          
          quando a condição for falsa

  **}**





```r
## Nesse caso salvamos em um objeto o numero 5 e pedimos para o R validar se o numero salvo no x é maior que 3
## x > 3
## Se esse teste der verdadeiro então ele printa o objeto na tela
## Caso não for ele retorna uma mensagem: O numero é menor que 3

x = 5

if(x > 3){ print(x)
          
  }else{
  
  print("O numero é menor que 3")
  
}


### mudando o x

x = 2

if(x > 3){ print(x)
          
  }else{
  
  print("O numero é menor que 3")
  
}
```

Veja que essas funções não podem ser aplicadas dentro de uma linha de um dataframe, para isso temos uma função que faz isso:

* **ifelse(teste_logico, resultado_true)**

Para isso precisamos criar uma coluna nova no nosso dataframe para receber essa função. Para isso utilizamos a seguinte sintaxe:

* **dataframe_nome_coluna_nova = O QUE A COLUNA VAI RECEBER**

```r
## VAMOS CRIAR UM COLUNA PARA REALIZAR NOSSO TESTE LOGICO
## Sempre que formos selecionar ou falar de colunas utilizamos o $
## Vamos pedir para o R criar a coluna logica e preencher a mesma com a informação UP caso o valor da coluna log2 seja maior que 0 e com DOWN, caso seja menor


df$tipo = ifelse(df$log2 > 0, "UP", "DOWN")

head(df,10)
```

* Nos comandos de repetições informamos ao R o que fazer enquanto determinada condição for verdadeira. Para isso utilizamos ou o **for** ou **while**.

**for(variavel/lista/teste_logico){**
  comandos para rodar
**}**


**while(variavel/lista/teste_logico){**
  comandos para rodar
**}**




```r
## Aqui no caso vamos dizer para o R realizar uma contagem e printar na nossa tela a sequencia numerica enquanto a condição i <= 10 for verdadeira.

i = 1

while(i <= 10){
  
  print(i)
  
  i = i+1
  
}



for(x in 1:10){
  print(x)
}
```

## Pacote dplyr

dplyr é um pacote para  manipulação de dados, fornecendo um conjunto consistente de funções que ajudam a resolver os desafios mais comuns de manipulação de dados. É importante resaltar que cada linha da base representa uma observação.
Cada coluna da base representa uma variável.


Principais funções:


* **select(dataframe, colunas)** escolhe colunas
* **filter(dataframe, filtragem)** escolhe casos com base em seus valores
* **mutate(dataframe, nome_nova_coluna = valor)** adiciona novas variáveis 
* **summarise(função, dados)** reduz vários valores a um único resumo.
* **arrange(dataframe, colunas)** altera a ordem das linhas
* **group_by(dataframe, colunas)** altera a ordem das linhas

?tidyverse ##consultar documentação do pacote



Todas essas funções seguem as mesmas características:

* O input é sempre uma tabela e o output é sempre um tabela.
* Colocamos o tabela no primeiro argumento e o que queremos fazer nos outros argumentos.
* A utilização é facilitada com o emprego do operador %>%.

As principais vantagens de se usar o dplyr em detrimento das funções do R base são:

* Manipular dados se torna uma tarefa muito mais simples.
* O código fica mais intuitivo de ser escrito e mais simples de ser lido.
* O pacote dplyr utiliza C e C++ por trás da maioria das funções, o que geralmente torna o código mais eficiente.
* É possível trabalhar com diferentes fontes de dados, como bases relacionais (SQL) e data.table.


```r
## Selecionando colunas
df_1 = select(df, padj, tipo)

head(df_1, 10)

rownames(df) =  df$gene

## Filtrando dados

df_UP = filter(df_1, tipo == "UP")


df_filter_padj = filter(df_1, padj < 0.05 | tipo == "UP")

head(df_filter_padj, 10)


## Reordenando
df = arrange(df, pvalue)

head(df)

df = arrange(df, desc(pvalue))

head(df)


## Permite operações de grupo no conceito "dividir-aplicar-combinar"

## Vamos voltar para nossa base geral e vamos criar grupos tipo
tipo_by = group_by(df, tipo)

tipo_by

## Agora podemos pedir para o R fazer operações conforme os grupos para isso usamos a função sumarise

summarise(tipo_by, count(tipo_by))





```

## Unindo tabelas


Vimos que algumas funções nos auxiliam a unir tabelas, porém elas precisam ter a mesma dimensão. Quando as tabelas forem de tamanho diferentes outras funções nos ajudam a realziar esse trabalho:

* **tibble()** Função para criar tabela

* **inner_join(DATAFRAME1x, DATAFRAME2y, QUAL A COLUNA VAI SER USADA DE REFERENCIA)**
Retorna todas as linhas de x onde há valores correspondentes em y e todas as colunas de x e y. Se houver várias correspondências entre x e y, todas as combinações das correspondências serão retornadas.

* **left_join(DATAFRAME1x, DATAFRAME2y, QUAL A COLUNA VAI SER USADA DE REFERENCIA)**
Retornar todas as linhas de x e todas as colunas de x e y. As linhas x sem correspondência em y terão NA valores nas novas colunas. Se houver várias correspondências entre x e y, todas as combinações das correspondências serão retornadas.

* **right_join(DATAFRAME1x, DATAFRAME2y, QUAL A COLUNA VAI SER USADA DE REFERENCIA)**
retorna todas as linhas de y e todas as colunas de x e y. As linhas y sem correspondência em x terão NAvalores nas novas colunas. Se houver várias correspondências entre x e y, todas as combinações das correspondências serão retornadas.

* **full_join(DATAFRAME1, DATAFRAME2, QUAL A COLUNA VAI SER USADA DE REFERENCIA)**
Retorna todas as linhas e todas as colunas de x e y. Onde não houver valores correspondentes, retorna NA o que está faltando.


A filtragem de junções mantém os casos do data.frame à esquerda:

* **semi_join()**
Retorna todas as linhas de x onde existem valores correspondentes y, mantendo apenas as colunas de x.
Uma semi-junção difere de uma junção interna porque uma junção interna retornará uma linha de x para cada linha correspondente de y, onde uma semi-junção nunca duplicará as linhas de x.

* **anti_join()**
Retorna todas as linhas de x onde não há valores correspondentes y, mantendo apenas as colunas de x.


```r

df = inner_join(df, annot, by = c( "gene"="ensembl_gene_id")) 

head(df)


## Criando tabelas

tb1 <- tibble(esp = c("esp 1", "esp 2", "esp 3"), 
              qtd = c(10, 3, 6)) 


tb2 <- tibble(nome = c("esp 1", "esp 2", "esp 3", "esp 4"), 
              data = c("2017-02-23", "2019-01-30", "2018-12-10", NA)) 

head(tb1, 6)
head(tb2, 6)


## Agora veremos os resultados da uniao

inner_join(tb2, tb1, by = c("nome"="esp")) 

full_join(tb1, tb2, by = c("esp" = "nome"))

left_join(tb2, tb1, by = c("nome"="esp"))

right_join(tb2, tb1, by = c("nome"="esp"))

semi_join(tb2, tb1, by = c("nome"="esp"))
anti_join(tb2, tb1, by = c("nome"="esp"))
```
